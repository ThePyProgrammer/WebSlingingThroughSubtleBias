```python
import pandas as pd
import requests
```

```python
template = "https://www.comichron.com/monthlycomicssales/{year:4d}/{year:4d}-{month:02d}.html"
```

```python
for year in range(2000, 2023):
    for month in range(1 if year != 2000 else 9, 13):
        print(year, month)
        url = template.format(year=year, month=month)
        df = pd.read_html(requests.get(url).content)[0]
        print(df[df["Comic-book Title"].str.lower().str.contains("ultimate") & df["Comic-book Title"].str.lower().str.contains("spider")])
```
